Data Cleaning

```python
# Check for missing values
print(menu_df.isnull().sum())
print(sales_df.isnull().sum())

# Check for duplicates
print(menu_df.duplicated().sum())
print(sales_df.duplicated().sum())

# Remove duplicates from menu_df and sales_df(if any)
menu_df.drop_duplicates(inplace=True)
sales_df.drop_duplicates(inplace=True)

# Handle missing values(if any).
menu_df.dropna(inplace=True)
sales_df.dropna(inplace = True)

print(menu_df.info())
print(sales_df.info())
```

Exploratory Data Analysis(EDA)

i) Menu Data Analysis

a) Calories distribution

```python
counts, bins= np.histogram(menu_df['Calories'], bins=20)

# Create a colormap with 20 distinct colors
colors=plt.cm.viridis(np.linspace(0,1,len(counts)))

# Plot each bar with a different color
plt.figure(figsize=(8, 6))
for count, x, color in zip(counts, bins[:-1], colors):
  plt.bar(x,count,width=np.diff(bins)[0], color=color,edgecolor='black')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

b) Total Fat vs Calories Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(menu_df['Total Fat'], menu_df['Calories'], color='purple')
plt.title('Total Fat vs Calories Scatter Plot')
plt.xlabel('Total Fat')
plt.ylabel('Calories')
plt.show()
```

c) Calories from Fat Distribution

```python
#Boxplot for Calories from Fat
plt.figure(figsize=(8,6))
sns.boxplot(x=menu_df['Calories from Fat'])
plt.title('Calories from Fat Distribution')
plt.xlabel('Calories from Fat')
plt.show()
```

d) Correlations Heatmap

```python
#Filter for numeric columns only
numeric_menu_df=menu_df.select_dtypes(include=['float64','int64'])

# Calculate the correlation matrix
correlation_matrix=numeric_menu_df.corr()

# Create the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Menu Data')
plt.show()
```

ii) Retail Sales Dataset Analysis

a) Sales Distribution


```python
#Histogram of Sales
#Calculate histogram values
counts, bins= np.histogram(sales_df['Total Amount'],bins=20)

#Create a colormap withh 20 distinct colors
colors=plt.cm.viridis(np.linspace(0,1,len(counts)))

#Plot each bar with a different color
plt.figure(figsize=(8,6))
for count, x, color in zip(counts, bins[:-1], colors):
  plt.bar(x,count,width=np.diff(bins)[0], color=color,edgecolor='black')
plt.title('Distribution of Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

b) Price vs Quantity Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(sales_df['Price per Unit'], sales_df['Quantity'], color='orange')
plt.title('Price vs Quantity Scatter Plot')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.yticks(np.round(plt.yticks()[0]))

plt.grid(alpha=0.5)  # Adding a light grid for better readability
plt.show()
```

c) Sales by Category barplot


```python
plt.figure(figsize=(8,6))
sns.barplot(x='Product Category', y='Total Amount', hue='Product Category',data=sales_df,palette='Set1',legend=False)
plt.title('Sales by Category')
plt.grid(alpha=0.2)
plt.show()
```

d) Heatmap for numerical feature Correlation

```python
#Filter for numeric columns only
numeric_sales_df = sales_df.select_dtypes(include=['float64','int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_sales_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of Numerical Feature Correlations")
plt.show()
```

TIME SERIES ANALYSIS

1. Retails Sales Data

```python
# Convert the 'Date' column to datetime objects
sales_df['Date'] = pd.to_datetime(sales_df['Date'])
```

```python
#obtaining a plot of Total Amount of particular dates with Seaborn
plt.figure(figsize=(15,5))
sns.lineplot(x='Date',y='Total Amount',data=sales_df)
plt.title('Total Amount Over Time')
plt.ylabel('Total Amount')
plt.show()
```

Time Series Analysis by Month

```python
#Creating Month and Date Columns
sales_df['Month'] = sales_df['Date'].dt.month
sales_df['Year'] = sales_df['Date'].dt.year

#calculating Total Monthly Sales

Monthly_sales=sales_df.groupby(['Year','Month'],as_index=False)["Total Amount"].sum()
sales_df['Monthly_sales']=Monthly_sales['Total Amount']
```

```python
#A plot of total_monyhly_sales throughout the year

plt.figure(figsize=(8,6))
sns.lineplot(x='Date',y='Monthly_sales',data=sales_df)
plt.title('Total Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.grid(alpha=0.5)
plt.show()
```

```python
#Deleting the month,year and Monthly_sales columns for Customer and Product Analysis
sales_df.drop(['Month','Year','Monthly_sales'],axis=1,inplace=True)
sales_df.head()
```

Customer and Product Analysis

```python
#creating age brackets

age_bins=[10,20,30,40,50,60]
age_labels=['10-19','20-29','30-39','40-49','50+']
sales_df['Age_Group']=pd.cut(sales_df['Age'],bins=age_bins,labels=age_labels,right=False)

#calculating Total Amount Spent by each Age_Group
total_amount_by_age=sales_df.groupby('Age_Group')['Total Amount'].sum()
print(total_amount_by_age)

#Grouping by age group and product category, and calculating the sum of spending
Grouped_data=sales_df.groupby(['Age_Group','Product Category'])['Total Amount'].sum().reset_index()
print(Grouped_data)
```

```python
#creating a bar plot to visualise the spending pattern
plt.figure(figsize=(8,6))
sns.barplot(x='Age_Group',y='Total Amount',hue='Product Category',data=Grouped_data)
plt.title('Total Amount Spent by Age Group and Product Category')
```

GENDER SPENDING ANALYSIS

```python
#Ccalculating Total amount spent by each Gender
total_amount_by_gender=sales_df.groupby('Gender')['Total Amount'].sum().reset_index()
plt.figure(figsize=(5,5))
plt.bar(total_amount_by_gender['Gender'],total_amount_by_gender['Total Amount'],color=['pink','black'])
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Total Amount Spent by each Gender')
```

Gender Product Preference

```python
Gender_preference=sales_df.groupby(['Gender','Product Category'])['Total Amount'].sum().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='Gender',y='Total Amount',hue='Product Category',data=Gender_preference)
plt.title('Gender Product Preference')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
```

```python
#Total Amount spent by the Average male and female

avg_amount_by_gender=sales_df.groupby('Gender')['Total Amount'].mean()
print(avg_amount_by_gender)
```
